* Higher the loan amount higher is the chances of getting accepted.
* Loan applications with Higher interest rates are mostly accepted (greater than 14-15%)
* Applications having high loan_percent_income (greater than 0.3) are more likely to be accepted than being rejected
* Applications of person having default on file is more likely to be accepted (around 45%).

## Feature Engineering

#### Removing unimportant feature

```python
# let's remove unimportant features from the dataset. But we will store this into a new dataset
new_df = df_train.drop(columns= ['id', 'person_age', 'cb_person_cred_hist_length'])
```

```python
new_df.head()
```

#### Converting Categorical data to numeric

```python
new_df = pd.get_dummies(new_df, drop_first=True, dtype='int')
```

```python
new_df.head()
```

```python
# Checking the correlation of different features with respect to loan status
new_df.corr()['loan_status'].sort_values(ascending=False).plot(kind='bar')
plt.xticks(rotation=90, fontsize=7)
plt.show()
```

## Model Training

```python
# Splitting the dataset into training set and testing set
X = new_df.drop(columns=['loan_status'])
y = new_df['loan_status']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

```python
print('X_train.shape: ', X_train.shape)
print("X_test.shape: ", X_test.shape)
print("Y_test.shape", y_test.shape)
print("Y_train.shape", y_train.shape)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(X_train, y_train)
```

```python
print("Score on training set: ", lg.score(X_train, y_train))
```

```python
from sklearn.metrics import accuracy_score
y_pred = lg.predict(X_test)
print("Score on test set: ", round(accuracy_score(y_test, y_pred)*100, 2))
```

```python
#Let's tune the hyperparameters of our model
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
params = {
    'max_iter':[100, 300, 500],
    'C':[1.0, 0.01, 0.1],
    'penalty':['l2', 'l1', None]
}
lg2 = LogisticRegression()
grid = GridSearchCV(lg2, params)
grid.fit(X_train_scaled, y_train)
```

```python
print("Best parameters: ", grid.best_params_)
```

```python
y_pred2 = grid.predict(X_test_scaled)
print("Accuracy on test set: ", round(accuracy_score(y_test, y_pred2)*100, 2))
```
